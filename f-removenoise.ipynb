```python
import noisereduce as nr
import librosa
import soundfile as sf

# Charger l'audio
audio, sr = librosa.load("/kaggle/input/daic-woz/357_P/357_AUDIO.wav", sr=None)

# Appliquer la réduction du bruit avec des paramètres ajustés
reduced_audio = nr.reduce_noise(
    y=audio,
    sr=sr,
    prop_decrease=0.8,  # Augmenter la réduction du bruit
    time_mask_smooth_ms=50,  # Lisser le masque temporel
    freq_mask_smooth_hz=200  # Lisser le masque fréquentiel
)

# Sauvegarder le fichier audio nettoyé
sf.write("audio_reduit.wav", reduced_audio, sr)
```

```python
import noisereduce as nr
import librosa
import soundfile as sf

# Charger l'audio
y, sr = librosa.load("/kaggle/input/daic-woz/357_P/357_AUDIO.wav", sr=None)

# Réduction de bruit douce, adaptée à l'analyse émotionnelle
reduced = nr.reduce_noise(
    y=y,
    sr=sr,
    stationary=False,      # meilleur pour les voix naturelles
    prop_decrease=0.7      # réduit modérément le bruit
)

# Sauvegarder le résultat
sf.write("audio_reduit_emo.wav", reduced, sr)
```
